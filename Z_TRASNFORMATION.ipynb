```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = np.exp(-0.2)  # Exponential decay base (e^(-0.2))
n = np.arange(0, 1000)  # Time indices for n >= 0

# Signal
def x_n(n, a):
    """
    Generate the signal e^(-0.2n) * u[n].

    Parameters:
        n (array): Discrete time indices.
        a (float): Base of the exponential decay.

    Returns:
        signal (array): Signal values for e^(-0.2n) * u[n].
    """
    return (a ** n) * (n >= 0)

x_values = x_n(n, a)

# Z-Transformation
def Z_Transformation(signal, n):
    """
    Compute the Z-Transform of the given signal.

    Parameters:
        signal (array): Input signal values.
        n (array): Discrete time indices.

    Returns:
        Z (array): Z-Transform values.
        roc (str): Region of Convergence (ROC).
    """
    N = len(signal)
    Z = np.zeros(N, dtype=complex)

    for k in range(N):
        Z[k] = np.sum(signal * np.exp(-1j * 2 * np.pi * n * k / N))

    # Determine the ROC based on the decay rate
    roc = f'ROC: |Z| > {1 / a}'  # ROC is |Z| > 1/a for causal signals

    return Z, roc

# Compute Z-Transform and ROC
Z, roc = Z_Transformation(x_values, n)

# Print the results
print(roc)

# Plot Magnitude and Phase for |z| = 1
theta = np.linspace(0, 2 * np.pi, 1000)  # Angles for the unit circle
z_circle = np.exp(1j * theta)  # Points on the unit circle
X_z = z_circle / (z_circle - a)  # Z-Transform on the unit circle

# Plot magnitude
plt.figure()
plt.plot(theta, np.abs(X_z))
plt.title("Magnitude of Z-Transform (|z| = 1)")
plt.xlabel("Angle (radians)")
plt.ylabel("Magnitude")
plt.grid(True)

# Plot phase
plt.figure()
plt.plot(theta, np.angle(X_z))
plt.title("Phase of Z-Transform (|z| = 1)")
plt.xlabel("Angle (radians)")
plt.ylabel("Phase (radians)")
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Signal Parameters and Setup
# -----------------------------
alpha = 0.2
a = np.exp(-alpha)  # This is e^(-0.2)
n = np.arange(0, 1000)  # Time index

# Define the signal: x[n] = e^(-0.2*n) * u[n]
def x_n(n, a):
    # (n>=0) acts as the unit step u[n]
    return (a ** n) * (n >= 0)

# Compute the signal values
x_values = x_n(n, a)

# ----------------------------------
# Z-Transformation (DFT-based method)
# ----------------------------------
def Z_Transformation(signal, n):
    N = len(signal)
    Z = np.zeros(N, dtype=complex)
    # Compute the "Z-transform" via summation over the finite-length signal.
    # This is equivalent to computing the DFT, which gives X(z) on the unit circle (z = e^(jθ)).
    for k in range(N):
        Z[k] = np.sum(signal * np.exp(-1j * 2 * np.pi * n * k / N))
    
    # In this template, 'roc' is computed in a placeholder way.
    roc = 'Entire Z-Plane'
    if np.abs(Z).any():
        roc = {np.max(np.abs(Z))}
    return Z, roc

# Compute the Z-transform of the signal using the above function.
Z, roc_computed = Z_Transformation(x_values, n)

# ----------------------------------
# Theoretical ROC
# ----------------------------------
# For x[n] = e^(-0.2n) u[n], the Z-transform converges for |z| > e^(-0.2).
roc_theoretical = a  # Since a = e^(-0.2)
print(f"Theoretical ROC: |z| > {roc_theoretical:.4f}")

# (Optional) Print a few values of the computed Z-transform:
print("First 10 values of the computed Z-Transformation:")
print(Z[:10])

# ----------------------------------
# Plotting on the Unit Circle
# ----------------------------------
# The computed Z (via our DFT-based function) corresponds to z = e^(jθ) for θ in [0, 2π).
theta = np.linspace(0, 2 * np.pi, len(Z), endpoint=False)

plt.figure(figsize=(10, 8))

# Plot magnitude
plt.subplot(2, 1, 1)
plt.plot(theta, np.abs(Z), 'b', linewidth=2)
plt.title('Magnitude of $X(z)$ on the Unit Circle ($z = e^{j\\theta}$)')
plt.xlabel('$\\theta$ (radians)')
plt.ylabel('$|X(z)|$')
plt.grid(True)

# Plot phase
plt.subplot(2, 1, 2)
plt.plot(theta, np.angle(Z), 'r', linewidth=2)
plt.title('Phase of $X(z)$ on the Unit Circle ($z = e^{j\\theta}$)')
plt.xlabel('$\\theta$ (radians)')
plt.ylabel('Phase (radians)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of ω from 0 to 2π
omega = np.linspace(0, 2 * np.pi, 1000, endpoint=False)
z = np.exp(1j * omega)  # z values on the unit circle

a = np.exp(-0.2)  # e^{-0.2}
z_inv = 1 / z     # z^{-1} = conjugate(z) since |z|=1

# Compute Z-Transform
X = 1 / (1 - a * z_inv)

# Magnitude and Phase
magnitude = np.abs(X)
phase = np.angle(X)  # Phase in radians between -π and π

# Plotting
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(omega, magnitude, label='Magnitude')
plt.title('Magnitude of the Z-Transform on the Unit Circle')
plt.xlabel('ω (radians)')
plt.ylabel('|X(z)|')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(omega, phase, label='Phase', color='orange')
plt.title('Phase of the Z-Transform on the Unit Circle')
plt.xlabel('ω (radians)')
plt.ylabel('Phase (radians)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
