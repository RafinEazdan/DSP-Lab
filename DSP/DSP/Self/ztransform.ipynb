```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a = 0.8
n = np.linspace(0, 50, 100)

def x_n(a,n):
    return (a**n) * (n>=2)

x = x_n(a,n)
plt.figure(figsize = (12,9))
plt.title("Signal")
plt.plot(n,x)
plt.xlabel("Time")
plt.ylabel("x(n)")
plt.show()
```

```python
def Z_trans(signal, n):
    N = len(signal)
    Z = np.zeros(N, dtype=complex)
    for k in range(N):
        Z[k] = np.sum(signal * np.exp(-1j * 2 * np.pi*n*k/N))

    return Z

Z = Z_trans(x,n)
roc = a
print(f"ROC: |z| > {a}")

fig, ax = plt.subplots()
circle = plt.Circle((0,0), radius = roc, edgecolor='r')
ax.add_patch(circle)
plt.xlim(-roc-1, roc+1)
plt.ylim(-roc-1, roc+1)
plt.grid(True, linestyle='--', linewidth=0.5)

plt.show()
    
```
