```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = np.exp(-0.2)
n = np.arange(0, 1000)
def x_n_e(n, a):
    return (base ** n) * (n >= 0)

x_values_e = x_n_e(n, base)

#Z-Transformation
def Z_Transformation(signal, n):
  N = len(signal)
  Z = np.zeros(N, dtype=complex)

  for k in range(N):
    Z[k] = np.sum(signal * np.exp(-1j * 2 * np.pi * n * k / N))

  roc = 'Entire Z-Plane'
  if np.abs(Z).any():
    roc = f'ROC: |Z| > {base}'

  return Z, roc
Z_e,roc_e = Z_Transformation(x_values_e, n)
print(roc_e)
print(Z_e)
```

```python
# Plotting ROC
fig, ax = plt.subplots()
circle = plt.Circle((0, 0), radius=np.max(np.abs(Z_e)), edgecolor='r')
ax.add_patch(circle)

plt.xlim(-np.max(np.abs(Z_e))-1, np.max(np.abs(Z_e))+1)
plt.ylim(-np.max(np.abs(Z_e))-1, np.max(np.abs(Z_e))+1)

plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Region Of Convergence')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```
